## **Country OLS Regression**

#### **Brunei**

```r
#Load climate data
brn <- read.csv("https://raw.githubusercontent.com/aris-zoleta/climate_ASEAN/main/brn.csv")
```

```r
#Declare our data to be a panel data set
brn.p <- pdata.frame(brn,index=c("year","country"))
head(brn.p)
```

#### Agricultural Production

```r
#fit simple linear regression model
model_brn <- lm(log(prod)~labor+log(land)+ log(prec) + log(CO2) + log(mean_temp), data=brn.p)

#view model summary
summary(model_brn)
```

# Panel Regression Country Specific Warming

```r
#Load climate data
climate <- read.csv("https://raw.githubusercontent.com/aris-zoleta/climate_ASEAN/main/raw_data_climate.csv")
```

```r
#Declare our data to be a panel data set
climate.p <- pdata.frame(climate,index=c("year","country"))
head(climate.p)
```

```r
#fixed effects Current Model
fullsample_FE <- plm(log(prod)~ labor + log(land)+ log(prec) + log(CO2) + log(mean_temp),data=climate.p,model="random")
stargazer(fullsample_FE,type='text')
```

```r
#fixed effects / within
fullsample_FE <- plm(log(prod)~ labor + lag(log(land))+ lag(log(prec)) + lag(log(CO2)) + lag(log(mean_temp)),data=climate.p,model="within")
stargazer(fullsample_FE,type='text')
```

```r
#fixed effects / within
fullsample_random <- plm(log(prod)~ labor + lag(log(land))+ lag(log(prec)) + lag(log(CO2)) + lag(log(mean_temp)),data=climate.p,model="random")
stargazer(fullsample_random ,type='text')
```

```r
#Run a hausman test comparing random and fixed effects
full_test <- phtest(fullsample_FE,fullsample_random)
print(full_test)
```

### Crop Production

```r
#fixed effects / within Cassava
crops_maize <- plm(log(maize)~ labor + lag(log(land))+ lag(log(prec))  + lag(log(mean_temp)),data=climate.p,model="random")
stargazer(crops_maize ,type='text')
```

```r
#fixed effects / within Cassava
crops_rice <- plm(log(rice)~ labor + lag(log(land))+ lag(log(prec))  + lag(log(mean_temp)),data=climate.p,model="random")
stargazer(crops_rice ,type='text')
```
